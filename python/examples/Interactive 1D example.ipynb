```python
from ipywidgets import interact
import matplotlib.pyplot as plt
import numpy as np
import goofit
```

```python
x = goofit.Observable('x', 0., 10.)
mean = goofit.Variable('mean', 1, -10, 10)
sigma = goofit.Variable('sigma', 1, 0, 5)
gauss = goofit.LandauPdf('gauss', x, mean, sigma)
```

```python
grid, pts = gauss.evaluatePdf(x)
gauss.setData(grid)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
v = ax.plot(grid.to_numpy().flatten(), pts, linewidth=2)[0]
ax.set_xlabel('xvar')
ax.set_ylabel('Normalized probability');
```

```python
@interact
def f(a=(0.0, 10.0), b=(0.0,10.0)):
    mean.value, sigma.value = a, b
    pts = gauss.evaluateAtPoints(x)
    v.set_ydata(pts)
```
