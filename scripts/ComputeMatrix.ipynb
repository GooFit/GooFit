```python
import re

import numpy as np
import sympy
from sympy.utilities.codegen import codegen
```

```python
M = sympy.symarray('M',shape=(5,5))
```

Zeroes:
* P1, 4pi
* P2, 4pi

```python
M[0,2] = 0
M[1,2] = 0
```

```python
Mi = sympy.Matrix(M).inv()
```

```python
Mcode = np.empty_like(M)
```

```python
res = re.compile(r'_result = (.*);', re.MULTILINE)
toarr = re.compile(r'M_(\d)_(\d)')
```

```python
for i in range(M.shape[0]):
    for j in range(M.shape[1]):

        out = Mi[i,j].simplify()
        [(_, c_code), (_, _)] = codegen(
             ("out", out), "C", "test", header=False)
        cout = res.findall(c_code)[0]
        Mcode[i,j] = toarr.sub(r'M(\1,\2)', cout)
```

```python
print('')
for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        print(f'O({i},{j}) = {Mcode[0,0]};\n\n')
```
